```python
import numpy as np
import pandas as pd
#import sklearn # This is a Machine Learning Algorithm
from sklearn.datasets import load_boston
df = load_boston()
```

```python
df.keys() # Returns all the keys of the dataset dictionary
```

```python
df
```

```python
print(df.DESCR) # Info about the dataset like on Coloums and Rows Info
```

```python
boston = pd.DataFrame(df.data, columns=df.feature_names)
boston.head() #Gives only first five rows of our data
```

```python
boston['MEDV'] = df.target
boston.head() # Gives only first five rows of our data with added MEDV column
```

```python
boston.isnull()
```

```python
boston.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split

X = boston.drop('MEDV', axis=1) # to seperate MEDV from real data
Y = boston['MEDV']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.15, random_state=5) #Here from every 100 set of datapoint 15 are selected for test_case

print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

### **Now let's import the Linear Regerssion model from sklearn and train it on the training dataset** 

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
## FITTING MODEL ON THE TRAINING DATASET

lin_model = LinearRegression() # Here Machine Learning Model is created

lin_model.fit(X_train, Y_train)
```

```python
# on Training set
y_train_predict = lin_model.predict(X_train)
rmse = (np.sqrt(mean_squared_error(Y_train, y_train_predict)))

print("The model performance for training set")
print('RMSE is {}'.format(rmse)) # Here error is calculated during training time 
print('\n')

# on testing set
y_test_predict = lin_model.predict(X_test)
rmse = (np.sqrt(mean_squared_error(Y_test, y_test_predict)))

print("The model performance for testing set")
print('RMSE is {}'.format(rmse)) # Here error is calculated during testing time
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(5,5))
plt.scatter(Y_test, y_test_predict)
```

```python
plt.figure(figsize=(5,5))
plt.scatter(Y_test, y_test_predict)
plt.plot([min(y_test_predict),max(y_test_predict)],[min(y_test_predict),max(y_test_predict)])
plt.xlabel('Actual')
plt.ylabel('Predicted')
```
