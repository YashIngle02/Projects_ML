```python
# Importing libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

# Loading Datasets

```python
# Loading datasets

iris = pd.read_csv('Iris.csv')
```

```python
# Displaying first five rows of dataset

iris.head()
```

## Data preparation

```python
# Checking info and describe 

print(iris.info())
iris.describe(include='all')
```

```python
# Checking count of null values for each row
iris.isnull().sum()
```

# Data Visualization

```python
# Visualization through Count_plot
plt.figure(figsize = (8, 8))
sb.countplot(data = iris, x='Species')
```

```python
plt.figure(figsize = (8, 8))
base_color = sb.color_palette('tab10')[1]
sb.countplot(data = iris, x='Species', color = base_color)

iris['Species'].value_counts()
```

```python
# Visualization through Violin_plot
plt.figure(figsize = (8, 8))
sb.violinplot(data = iris, x = 'Species', y = 'PetalWidthCm')
plt.show()
```

```python
# Visualization through Box_plot
plt.figure(figsize = (8, 8))
base_color = sb.color_palette('tab10')[8]
sb.boxplot(data = iris, x = 'Species', y = 'PetalWidthCm', color = base_color)
plt.xticks(rotation=15)
```

```python
# Calculating Average for PetalWidthCm
plt.figure(figsize = (8, 8))
iris_avg=iris["PetalWidthCm"].groupby(iris["Species"]).mean()
print(iris_avg)
iris_avg.plot.pie()
```

```python
# Calculating Average for PetalLengthCm
plt.figure(figsize = (8, 8))
iris_avg=iris["PetalLengthCm"].groupby(iris["Species"]).mean()
print(iris_avg)
iris_avg.plot.pie()
```

```python
# Calculating Average for SepalLengthCm
plt.figure(figsize = (8, 8))
iris_avg=iris["SepalLengthCm"].groupby(iris["Species"]).mean()
print(iris_avg)
iris_avg.plot.pie()
```

```python
# Calculating Average for SepalWidthCm
plt.figure(figsize = (8, 8))
iris_avg=iris["SepalWidthCm"].groupby(iris["Species"]).mean()
print(iris_avg)
iris_avg.plot.pie()
```

```python
# Histogram 
plt.figure(figsize = (8, 8))
x = iris["SepalLengthCm"]
  
plt.hist(x, bins = 30, color = "blue")
plt.title("Sepal Length in cm")
plt.xlabel("Sepal_Length_cm")
plt.ylabel("Count")
```

```python
# Dist_plot
plt.figure(figsize = (8, 8))
sb.distplot(iris['SepalLengthCm']);
```

```python
plt.figure(figsize = (8, 8))
sb.distplot(iris['SepalLengthCm'], kde=False);
```
